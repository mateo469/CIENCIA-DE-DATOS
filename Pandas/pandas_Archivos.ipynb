**Vamos aprender agrupar datos**


```python
#Crear mi nuevo grupo
grupo1 = archivo1.groupby('PROVINCIA')[['IMPORTE CLIENTE','IMPORTE PROVEEDOR']].mean().reset_index()
grupo1
```

```python
#Filtrar informacion de grupos de una sola columna
grupo_columns = archivo1.groupby('CONCEPTO').get_group('EQUIPOS')
grupo_columns
```

```python
#Mostrar informacion por grupos
info_group = archivo1.groupby('PROVINCIA').groups.items()
info_group
```

```python
#Filtrar informacion por cadenas de textos que contengas palabras coincidentes
archivo1[archivo1['CONCEPTO'].str.contains('EQUIP')]
```

**Crearcion de tablas dinamicas**


```python
#Mi primer tabla dinamica
tabla1 = archivo1.pivot_table(
    index='PROVINCIA',
    values=['IMPORTE CLIENTE','IMPORTE PROVEEDOR'],
    aggfunc=['sum']
)

tabla1
```
