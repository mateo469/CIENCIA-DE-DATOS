*¿Qué es Pandas?*

Pandas es una biblioteca de Python para el análisis y manipulación de datos. Ofrece estructuras de datos como Series y DataFrame, que hacen que trabajar con datos tabulares (como los de Excel o CSV) sea muy sencillo.

```python
#Crear un diccionario para un dataframe
informe = {
    'Id': [1,2,3,4,5,6,7,8],
    'Nombre':['Carlos', 'Maria', 'Salome', 'Isabel', 'Tomas', 'Nicole', 'Yamileth', 'Ana'],
    'Sexo':['M', 'F', 'F', 'F', 'M', '', 'F', 'F'],
    'Edad':[21, 19, 33, 44, 23, 16, 15, 45],
    'Estado':['Veracruz', 'Tabasco', 'Jalisco', 'Jalisco', 'Tabasco', 'Veracruz', 'Tabasco', 'Jalisco'],
    'Peso':[50, 60, 45, 70, 80, 65, 67, 66],
    'Estatura':[1.45, 1.55, 1.60, 1.65, 1.70, 1.75, 1.79, 1.80],
    'Calificacion':[100, 99, 76, 55, 69, 88, 95.0, 85.9]
}

#Covertir mi diccionario a un dataframe
dato = pd.DataFrame(informe).set_index('Id')
dato
#Guardar el dataframe creado en un archivo csv
#dato.to_excel('informe.xlsx', index=True, sheet_name='base')
```
